```python
title_list = []  # Initialize an empty list to store the titles

for i in range(1, 300):
    xpath = '//*[@id="listCardContainer"]/div[1]/div/article[' + str(i) + ']'
    element = driver.find_element(By.XPATH, xpath)
    
    title_element = element.find_element(By.CSS_SELECTOR, '.overflow-hidden.overflow-ellipsis.whitespace-nowrap.max-w-80pe.po\\:max-w-full')
    title = title_element.text
    
    title_list.append(title)  # Append each title to the list

# Print the titles
for title in title_list:
    print("Title:", title)
```

```python
rent_list = []  # Initialize an empty list to store the rents

for i in range(1, 300):
    xpath = '//*[@id="listCardContainer"]/div[1]/div/article[' + str(i) + ']'
    element = driver.find_element(By.XPATH, xpath)
    
    rent_element = element.find_element(By.ID, 'minimumRent')
    rent = rent_element.text
    
    rent_list.append(rent)  # Append each rent to the list

# Print the rents
for rent in rent_list:
    print("Rent:", rent)
```

```python
print(len(rent_list))
```

```python
deposit_list = []  # Initialize an empty list to store the deposits

for i in range(1, 300):
    xpath = '//*[@id="listCardContainer"]/div[1]/div/article[' + str(i) + ']'
    element = driver.find_element(By.XPATH, xpath)
    
    deposit_element = element.find_element(By.ID, 'roomType')
    deposit = deposit_element.text
    
    deposit_list.append(deposit)  # Append each deposit to the list

# Print the deposits
for deposit in deposit_list:
    print("Deposit:", deposit)
```

```python
print(len(deposit_list))
```

```python
builtup_list = []  # Initialize an empty list to store the built-up areas

for i in range(1, 300):
    xpath = '//*[@id="listCardContainer"]/div[1]/div/article[' + str(i) + ']'
    element = driver.find_element(By.XPATH, xpath)
    
    builtup_element = element.find_element(By.ID, 'unitCode')
    builtup = builtup_element.text
    
    builtup_list.append(builtup)  # Append each built-up area to the list

# Print the built-up areas
for builtup in builtup_list:
    print("Built-up Area:", builtup)
```

```python
print(len(builtup_list))
```

```python
furnishing_list = []  # Initialize an empty list to store the furnishing information

for i in range(1, 300):
    xpath = '//*[@id="listCardContainer"]/div[1]/div/article[' + str(i) + ']'
    element = driver.find_element(By.XPATH, xpath)
    
    parent_div = element.find_element(By.CLASS_NAME, 'flex-1.border-r.border-r-solid.border-r-cardbordercolor.flex')
    furnishing_element = parent_div.find_element(By.CLASS_NAME, 'font-semibold')
    furnishing = furnishing_element.text
    
    furnishing_list.append(furnishing)  # Append each furnishing information to the list

# Print the furnishing information
for furnishing in furnishing_list:
    print("Furnishing:", furnishing)
```
